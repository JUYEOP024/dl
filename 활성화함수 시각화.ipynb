Sigmoid() 활성화함수: 이진 분류의 출력층에서 주로 사용

```python
import torch
import torch.nn as nn
from matplotlib import pyplot as plt
```

```python
torch.sort(torch.randn(100)* 10)
```

- torch.randn(100) : 평균 0, 표준편자 1의 정규분포에서 100개의 랜덤값
- *10 : 평균 0, 표준편차 10인 정규분포의 랜덤값 100개
- sort()[0] : 오름차순 정렬한 후에 , 정렬된 텐서만 반환[0]은 값.[1]은 인덱스

```python
x = torch.sort(torch.randn(100)* 10)[0] 

x
```

```python
act = nn.Sigmoid() # 객체 지향 방식으로 정의해서 사용
print(act(x))
```

```python
print(torch.sigmoid(x)) # 함수 호출 방식으로 사용
```

```python
plt.plot(x.numpy(), torch.sigmoid(x).numpy())

plt.show()
```

tanh()활성화 함수: 순환 신경망(RNN)등 시계열 모델에서 중심화된 출력값이 필요할때 사용

```python
# 객체지향방식으로 사용
act = nn.Tanh()
print(act(x))
```

```python
# 함수 호출 방식
torch.sigmoid(x)
```

```python
plt.plot(x.numpy(), torch.tanh(x).numpy())

plt.show()
```

ReLU () 활성화 함수 : 대부분의 DNN (Deep Neural Network) 활성화 계층에서 기본값으로 사용한다.


```python
# 객체 지향 방식
act = nn.ReLU()
act(x)
```

```python
# 함수 형태
torch.relu(x)
```

```python
plt.plot(x.numpy(), torch.relu(x).numpy())

plt.show()
```

음수에 대해서는 Dead ReLU 현상이 발생한다

Leaky ReLU () 활성화 함수 : ReLU의 한계(죽은 뉴런)을 보완하기 위해 나온 활성화 함수

```python
act = nn.LeakyReLU(negative_slope=0.1)  # 음수일때도 완전히 0으로 죽지않고, 10% 만큼 누출시켜 계산함 = alpha
act(x)
```

```python
# 함수 호출 방식
torch.nn.functional.leaky_relu(x, negative_slope=0.1)
```

```python
plt.plot(x.numpy(), act(x).detach().numpy())
plt.show()
```
